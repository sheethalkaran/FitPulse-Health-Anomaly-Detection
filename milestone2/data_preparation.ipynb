```python
import pandas as pd

df = pd.read_csv("/content/drive/MyDrive/clean_fitness_data.csv")
df["DateTime"] = pd.to_datetime(df["DateTime"])
```

```python
df["TotalMinutesAsleep"] = df["TotalMinutesAsleep"].fillna(0)
```

```python
# Select 5 unique users
selected_ids = df["Id"].unique()[:5]

df_reduced = (
    df[df["Id"].isin(selected_ids)]
    .sort_values("DateTime")
    .groupby("Id")
    .head(15)
    .reset_index(drop=True)
)
```

```python
save_path = "/content/drive/MyDrive/FitPulse/Milestone2/data/"
import os
os.makedirs(save_path, exist_ok=True)

df_reduced.to_csv(
    save_path + "milestone2_reduced_fitness_data.csv",
    index=False
)
```
